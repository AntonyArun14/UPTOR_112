```python
def price_per_carat(price, carat):
    return price / carat if carat != 0 else 0

diamonds_df1['price_per_carat'] = diamonds_df1.apply(lambda row: price_per_carat(row['price'], row['carat']), axis=1)
diamonds_df1
```
